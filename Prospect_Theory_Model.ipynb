# **Prospect Theory**

A theory based on the assumption that losses and gains are valued differently, allowing individuals to make decisions based on gains or losses. Developed by economist Daniel Kahneman and mathematical psychologist Amos Tversky in 1979.

This is also known as the loss-aversion theory.

This theory is part of behavioral economics, a branch of economics that is an intersection between economics and psychology. The section is "grounded in empirical observations of human behavior, which have demonstrated that people do not always make what neoclassical economists consider the 'rational' or 'optimal' decision, even if they have the information and the tools available to do so" (University of Chicago).

The loss-aversion theory gathers bounded rationality and bounded willpower when making decisions based on gains and losses. It encounters how individuals will rather choose gains based on the emotional impacts it can leave rather than a loss.

# Model
There are two functions that make up this model.

<br>

The value function is defined as:

$$ V(x)=   \left\{
\begin{array}{ll}
      x^{\alpha} & \text{if } x \geq 0 \\
      -\lambda(-x)^{\beta} & \text{if } x \leq 0 \\
\end{array}
\right.  $$

<br>

The probability weighting function is defined as:
$$
W(p) = \frac{p^{\gamma}}{(p^{\gamma} + (1-p)^{\gamma})^{1/\gamma}}
$$ \\

Meaning of Symbols \\
$ {\alpha} $ = Curvature for gains \\
$ {\beta} $ = Curvature for losses \\
$ {\gamma} $ = Loss aversion, the tendency for individuals to weigh losses more heavily than gains of the same magnitude

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = 0.50
beta = 0.50
lambda_ = 2.50

x_values = np.linspace(-100, 100, 1000)

def value_function(x, alpha, beta, lambda_):
    return np.where(x >= 0, x**alpha, -lambda_ * (-x)**beta)


plt.figure(figsize=(8, 6))
plt.plot(x_values, value_function(x_values, alpha, beta, lambda_), label='Value Function', color='blue')
plt.xlabel('Outcome')
plt.ylabel('Value')
plt.title('Prospect Theory Model')
plt.legend()
plt.grid(True)
plt.show()
```

Individuals will try to be risk seeking and risk averse (based on the concave and convex value functions).

# Using Prospect Theory to Understand Consumer Choices with Insurance

Looking at the preferred level of risk per person, people will buy insurance and pay a deductible to recieve some amount of coverage. Those that are risk averse will pay the premium and gain some level of coverage if an accident were to occur.

As the weighed probability of an accident increases the people that are more likely to purchase insurance to avoid a large potential loss weigh the probability of an accident much higher than those that do not weigh that probability in the same way.

```python
alpha = 0.88
beta = 0.88
lambda_ = 2.25

# Insurance Plans
premium_insurance = 1000
coverage_limit = 20000
potential_loss = 20000

# Expected losses of insurance based on accidents
expected_loss_insurance = premium_insurance + (1/100) * coverage_limit
expected_loss_no_insurance = potential_loss

v_insurance = value_function(-expected_loss_insurance, alpha, beta, lambda_)
v_no_insurance = value_function(-expected_loss_no_insurance, alpha, beta, lambda_)

plt.figure(figsize=(8, 6))
plt.plot([0, 1], [0, v_insurance], label='Option A: Insurance', color='blue')
plt.plot([0, 1], [0, v_no_insurance], label='Option B: No Insurance', color='green')
plt.xlabel('Probability Weighting')
plt.ylabel('Value')
plt.title('Analysis of Insurance Decision')
plt.legend()
plt.grid(True)
plt.show()
```
